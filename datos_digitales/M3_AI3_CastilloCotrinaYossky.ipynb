**07-Captura el código html y cierra la página (si quieres la puedes dejar abierta para inspeccionar el código).**

```python
html_text = driver.page_source

html_text
```

**08-A partir del texto obtenido, convierte el texto a un árbol de nodos mediante html.document_fromstring().**

```python
tree = html.document_fromstring(html_text)

tree
```

**09-Captura el título y el precio de los productos comprados, habitualmente, a partir del árbol generado en el punto anterior. Puedes usar cualquier método de selección: CSSSelect, Xpath o incluso parseado de texto mediante expresiones regulares (a partir del texto obtenido en el punto 7).**

> Se informó al tutor que no se encontró tabla de precios de los productos comprados, por lo que me indicó obtener los datos de la sección de "Dirección, Elenco, Géneros, ...".

```python
title = tree.cssselect("h1[data-automation-id = title]")[0].text

print("Título:", title)

info = tree.cssselect("div#meta-info div")
dls = info[0].findall("dl")

arr = []

for dl in dls:
    col1 = dl.findall("dt")[0].text_content()    
    col2 = dl.findall("dd")[0].text_content()
    
    arr.append([col1, col2])
    
arr
```

**10-Presenta los resultados anteriores en formato tabla.**

```python
data_df = pd.DataFrame(arr, columns=["col1", "col2"])

data_df
```
